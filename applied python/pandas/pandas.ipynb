```python
# for changing the places of columns
df1[['State','Total day calls', 'Account length', 'Area code', 'International plan',
       'Voice mail plan', 'Number vmail messages', 'Total day minutes',
       
       'Total eve calls', 'Total eve charge', 'Total night minutes',
       'Total night calls', 'Total night charge', 'Total intl minutes',
       'Total intl calls', 'Total intl charge', 'Customer service calls',
       'Churn']]
```

```python
# loc()- location base indexing
# iloc()- integer base indexing
df1['Total intl charge']>1.46
```

```python
# for data frame of the particular elements
df1[df1['Total intl charge']>1.46]
```

```python
# for particular range 
(df1['Total intl charge']>1.46) & (df1['Total intl charge']>3.0)
```

```python
# for data frame
df1[(df1['Total intl charge']>1.46) & (df1['Total intl charge']<3.0)]
```

```python
# for copareing multiple multiple columns
df3=df1[(df1['Total intl charge']>1.46) & (df1['Total intl charge']<3.0) & (df1['International plan']=='Yes')]
df3
```

```python
# show the table whose internation palan is no and Total eve minutes is in betw 120 - 180
df4=df1[(df1['Total eve minutes']>120) & (df1['Total eve minutes']<180) & (df1['International plan']=='No')]
df4
```

```python
pd.concat([df3,df4],axis=0)
```
